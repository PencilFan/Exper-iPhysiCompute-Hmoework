```python
import numpy as np
import pandas as pd
H,N=50,500000#H是下落高度,相当于是纵向钉子的个数,N是下落的小球个数(或者说是实验次数)
MatSize=[H,N]
MatDeltaX=np.random.choice([-1,1],MatSize)#构造位移矩阵
MatDeltaX=pd.DataFrame(MatDeltaX)
XEndList=MatDeltaX.sum()#利用pandas模块可以很方便地算出每列位移之和
IndexList=XEndList.unique()#小球在底部的位置坐标
IndexList.sort()#将位置坐标排序
XUniqueList=XEndList.value_counts()#计算每一位置坐标所对应的小球个数
XUniqueList.sort_index(inplace=True)#将得到的数目列表排序
#TODO 作图
import matplotlib.pyplot as plt
plt.plot(IndexList,XUniqueList)
plt.bar(IndexList,XUniqueList)
plt.show()
```
