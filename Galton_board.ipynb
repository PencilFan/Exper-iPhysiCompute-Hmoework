与标准的整体分布函数比较scipy.stats.norm

```python
import numpy as np
import pandas as pd
H,N=200,500000#H是下落高度,相当于是纵向钉子的个数,N是下落的小球个数(或者说是实验次数)
MatSize=[H,N]
MatDeltaX=np.random.choice([-1,1],MatSize)
MatDeltaX=pd.DataFrame(MatDeltaX)
XEndList=MatDeltaX.sum()
IndexList=XEndList.unique()
IndexList.sort()
XUniqueList=XEndList.value_counts()#之后就可以用索引查找某一个元素的频率
XUniqueList.sort_index(inplace=True)
#TODO 作图
import matplotlib.pyplot as plt
plt.plot(IndexList,XUniqueList)
plt.bar(IndexList,XUniqueList)
plt.show()
```
