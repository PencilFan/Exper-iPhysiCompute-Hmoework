```python
import numpy as np
#TODO 初始化
G,M,m=1,1,1
N,tN=1000,50
r0,v0=[1,0],[0,2]#要和维数(dim/dimention)保持一致
dim=len(r0)
if dim!=len(v0):
    print('初始坐标与初始速度的维数不一致,请重新输入')
    exit#还是会执行下面的程序，如何修改？
#代码：若
tlist=np.linspace(0,tN,N)
deltat=tlist[1]-tlist[0]
r=np.zeros((N,dim))
v=np.zeros((N,dim))
r[0,:]=r0
v[0,:]=v0
k=G*M*m
#TODO 计算数据
for i in range(0,N-1):
    for n in range(dim):
        q=r[i,n]
        norm=np.linalg.norm(r[n,:])
        v[i+1,n]=-k*q*deltat/norm**3/m+v[i,n]#下一时刻的速度分量,注意负号表示吸引力
        deltaq=(v[i+1,n]+v[i,n])/2*deltat
        r[i+1,n]=q+deltaq#下一时刻的坐标分量
#TODO 作图
import matplotlib.pyplot as plt
plt.plot(r[:,0],r[:,1])#如何实现任意维度的作图
plt.show()
#稳定情况,作圆周运动的情况,被甩出去的情况？
#初始时不稳定,但当时间足够长后会达到稳定,此时稳定的椭圆半长轴与理论值是否相等?
```
